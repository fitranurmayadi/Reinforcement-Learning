```python
import gymnasium as gym
import time
from StarshipLanderEnv.envs import StarshipLanderEnv  # Pastikan path ini sesuai
from stable_baselines3.common.env_checker import check_env

# Buat environment dengan render_mode "human"
env = gym.make("StarshipLanderEnv-v0", render_mode="human")
check_env(env)  # Mengecek environment, opsional

# Reset environment untuk memulai episode baru
obs, info = env.reset()

# Loop untuk menjalankan satu episode hingga 5000 langkah
for step in range(5000):
    # Ambil aksi secara acak (atau ganti dengan aksi dari agent RL Anda)
    action = env.action_space.sample()
    obs, reward, done, truncated, info = env.step(action)
    
    # Render environment (render_mode "human" akan membuka GUI)
    env.render()
    
    # Delay kecil agar simulasi tidak berjalan terlalu cepat
    #time.sleep(0.001)
    
    # Cek apakah episode sudah selesai (baik karena landing sukses atau crash)
    if done or truncated:
        print(f"Episode selesai pada langkah ke-{step + 1}")
        break

env.close()
```

```python
import gymnasium as gym
from StarshipLanderEnv.envs import StarshipLanderEnv
from stable_baselines3 import SAC
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.callbacks import CheckpointCallback

# ================================
# Pelatihan dengan Vectorized Environment
# ================================

# Buat environment vectorized BipedalWalker-v3 (dengan 8 environment paralel)
env_train = make_vec_env("StarshipLanderEnv-v0", n_envs=8, vec_env_cls=DummyVecEnv)

# Definisikan hyperparameter SAC secara lengkap
sac_params = {
    "learning_rate": 5e-4,      # Kecepatan pembelajaran
    "buffer_size": 1000000,     # Kapasitas replay buffer
    "batch_size": 256,          # Ukuran batch untuk update
    "policy_kwargs": dict(net_arch=[512, 512]),  # Arsitektur jaringan
    "use_sde": True,            # Gunakan State-Dependent Exploration (SDE)
    "sde_sample_freq": 32,      # Update noise SDE setiap 64 langkah
}

# Buat model SAC dengan TensorBoard logging
model = SAC(
    "MlpPolicy",
    env_train,
    device="cuda",  # Gunakan GPU jika tersedia
    verbose=1,
    tensorboard_log="./sac_starship/",
    **sac_params
)

# Callback untuk menyimpan model secara berkala (setiap 10.000 timesteps)
checkpoint_callback = CheckpointCallback(
    save_freq=10000,
    save_path="./sac_starship/",
    name_prefix="sac_starship"
)

# Mulai pelatihan model selama 1.000.000 timesteps dengan progress bar
model.learn(total_timesteps=200000, progress_bar=True, callback=checkpoint_callback)

# Simpan model final
model.save("sac_starship")
print("Model telah disimpan sebagai sac_starship.zip")

# Tutup environment pelatihan
env_train.close()

# ================================
# Evaluasi Model dalam Mode Human
# ================================


```

```python
import gymnasium as gym
from StarshipLanderEnv.envs import StarshipLanderEnv
from stable_baselines3 import SAC
import time

# Buat environment dalam mode human untuk visualisasi
env = gym.make("StarshipLanderEnv-v0", render_mode="human")

# Muat model SAC yang telah disimpan
model = SAC.load("sac_starship.zip")

# Reset environment
obs, info = env.reset()

try:
    while True:
        # Dapatkan aksi dari model (deterministik untuk evaluasi)
        action, _ = model.predict(obs, deterministic=True)
        # Jalankan satu langkah dalam environment
        obs, reward, terminated, truncated, info = env.step(action)
        # Render environment
        env.render()
        # Jika episode selesai, reset environment
        if terminated or truncated:
            obs, info = env.reset()
        # Delay kecil agar rendering terlihat halus
        time.sleep(0.01)
except KeyboardInterrupt:
    print("Inferensi dihentikan oleh pengguna.")
finally:
    env.close()
```
